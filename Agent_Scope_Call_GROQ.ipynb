```python
import os
from dotenv import load_dotenv

from agentscope.agent import ReActAgent
from agentscope.message import Msg
from agentscope.model import OpenAIChatModel
from agentscope.formatter import OpenAIChatFormatter
from agentscope.memory import InMemoryMemory

# 1) carrega .env
ENV_PATH = r"C:\Users\fepac\Unicamp_Project\.env"
load_dotenv(ENV_PATH, override=True)

# 2) cria agente via Groq (OpenAI-compatible)
agent = ReActAgent(
    name="Friday",
    sys_prompt="You're a helpful assistant named Friday.",
    model=OpenAIChatModel(
        model_name="llama-3.3-70b-versatile",     # ✅ modelo novo suportado
        api_key=os.getenv("GROQ_API_KEY"),
        base_url=os.getenv("GROQ_BASE_URL"),      # https://api.groq.com/openai/v1
        stream=False,
    ),
    memory=InMemoryMemory(),
    formatter=OpenAIChatFormatter(),
    toolkit=None,
)

# 3) pergunta
msg = Msg("user", "Explique o que é ReAct em 3 linhas.", role="user")
reply = await agent(msg)
print(reply.get_text_content())
```
